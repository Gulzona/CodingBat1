```python
x = np.arange(0,4*np.pi, 0.1)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, linewidth=4)
plt.show()
```

```python
plt.plot(x,y, linewidth=4, linestyle=":")
plt.show()
```

```python
plt.plot(x,y, linewidth=4, linestyle=":", color="blue")
plt.show()
```

```python
plt.plot(x,y, linewidth=4, linestyle="--", color="blue", marker='v')
plt.title("sin(x) funksiya")
plt.show()
```

```python
 plt.plot(x,y, linewidth=4, linestyle="--", color="blue", marker='v')
plt.title("sin(x) funksiya")
plt.xlabel("pi")
plt.show()
```

```python
plt.plot(x,y, linewidth=4, linestyle="--", color="blue", marker='v')
plt.title("sin(x) funksiya")
plt.xlabel("pi")
plt.label("sin(x)")
plt.show()
```

```python
plt.plot(x,y, linewidth=4, linestyle="--", color="blue", marker='v')
plt.title("sin(x) funksiya")
plt.xlabel("pi")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(x,y, linewidth=4, linestyle="--", color="blue", marker='v')
plt.title("sin(x) funksiya")
plt.xlabel("pi")
plt.grid()
plt.show()
```
