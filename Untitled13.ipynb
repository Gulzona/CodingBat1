```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sine_recursive(x, d=0):
    if d == 5:
        return
    y = np.sin(x)
    plt.plot(x, y)
    plot_sine_recursive(x - np.pi/4, d+1)
    plot_sine_recursive(x + np.pi/4, d+1)
x = np.linspace(-10, 10, 1000)
plot_sine_recursive(x)
plt.xlim(-10, 10)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
def plot_cosine_recursive(x, d=0):
    if d == 5:
        return
    y = np.cos(x)
    plt.plot(x, y)
    plot_cosine_recursive(x - np.pi/4, d+1)
    plot_cosine_recursive(x + np.pi/4, d+1)

x = np.linspace(-10, 10, 1000)
plot_cosine_recursive(x)
plt.xlim(-10, 10)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
def plot_cosine_recursive(x, d=0):
    if d == 5:
        return
    y = np.cos(x)
    plt.plot(x, y, color='green')
    plot_cosine_recursive(x - np.pi/4, d+1)
    plot_cosine_recursive(x + np.pi/4, d+1)

x = np.linspace(-10, 10, 1000)

plt.plot(x, np.sin(x), color='red')
plot_cosine_recursive(x)

plt.xlim(-10, 10)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
def plot_tangent_recursive(x, d=0):
    if d == 5:
        return
    y = np.tan(x)
    plt.plot(x, y, color='blue')
    plot_tangent_recursive(x - np.pi/4, d+1)
    plot_tangent_recursive(x + np.pi/4, d+1)

x = np.linspace(-10, 10, 1000)

plt.plot(x, np.sin(x), color='red')
plot_tangent_recursive(x)

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```
